```python
import pandas as pd
import numpy as np
```

```python

data = {
    "Customer_ID": [101, 102, 103, 103, 104, 105, 106, 107],
    "Age": [25, np.nan, 28, 28, 40, 35, 29, 31],
    "Gender": ["Female", "Male", "Female", "Female", "Male", "Female", "Male", "Female"],
    "City": ["Hyderabad", "Chennai", "Hyderabad", "Hyderabad", "Bangalore", "Delhi", "Mumbai", "Chennai"],
    "Monthly_Spend": [2500, 3200, 3200, 3200, 4500, 1800, "3000", 2700],
    "Complaints": [1, 2, 2, 2, 15, 5, 1, np.nan],
    "Membership": ["Gold", "Silver", "Silver", "Silver", "Gold", "Bronze", "Silver", "Gold"],
    "Churn": ["No", "Yes", "Yes", "Yes", "No", "Yes", "No", "No"]
}

df = pd.DataFrame(data)

# Display the dataset
df
```

```python
df['Complaints'].fillna(df['Complaints'].median(),inplace=True)
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.drop_duplicates()
```

```python
# Convert Monthly_Spend to integer
df["Monthly_Spend"] = df["Monthly_Spend"].astype(int)
```

```python
df.info()
```

```python
Q1 = df["Complaints"].quantile(0.25) # This means 25% of the complaint values are less than or equal to 1.75.
Q3 = df["Complaints"].quantile(0.75) #This means 75% of the complaint values are less than or equal to 2.75.
print("Q1:", Q1)
print("Q3:", Q3)
IQR = Q3 - Q1 # 2.75 -1.75 =1

lower = Q1 - 1.5 * IQR #1.75-1.5*1
upper = Q3 + 1.5 * IQR #1.75+1.5*1

outliers = df[(df["Complaints"] < lower) | (df["Complaints"] > upper)]
#Show me every row where the complaint count is less than the lower limit OR greater than the upper limit.

outliers
```

# Business Insights

### Insight 1
The Age column had missing values, which were filled using the median to avoid losing valuable customer records.

### Insight 2
One duplicate customer record was found and removed to prevent biased analysis.

### Insight 3
The Monthly_Spend column was converted from string to integer so numerical calculations such as averages and correlations could be performed.

### Insight 4
A customer with 15 complaints was identified as a potential outlier. Since this could represent a genuinely dissatisfied customer, it should be investigated rather than removed immediately.

### Insight 5
Customers with higher complaint counts appear more likely to churn, suggesting that improving customer support could help reduce churn.
