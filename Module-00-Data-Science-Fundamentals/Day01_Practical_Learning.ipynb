## 3. Data Science Lifecycle
'''
 Business Problem
- Customers are leaving the company.

Collect Data
- Collect customer age, monthly spending, complaints and churn history.

EDA
- Explore relationships between complaints, spending and churn.

Feature Engineering
- Prepare useful features such as complaint frequency or encoded categorical values.

Model Building
- Train a classification model such as Logistic Regression or Decision Tree.

Evaluation
- Evaluate using Accuracy, Precision, Recall and F1-score.

Deployment
- Predict customers likely to churn and help the business take preventive actions.
'''

## 4. Business Recommendation

Customers with a high number of complaints appear more likely to leave. The company should:

- Prioritize complaint resolution.
- Improve customer support response time.
- Follow up with dissatisfied customers.
- Offer retention incentives for high-risk customers.
