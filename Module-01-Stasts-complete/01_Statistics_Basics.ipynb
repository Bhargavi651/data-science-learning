# Statistics Basics using Python

This notebook covers the fundamental statistical concepts required for Data Science.

Topics Covered:
- Mean
- Median
- Mode
- Range
- Variance
- Standard Deviation
- Quartiles
- Percentiles
- Outlier Detection (IQR)
- Correlation

```python
import pandas as pd
import numpy as np
```

```python
data = {"marks":[45,52,56,61,60,65,70,72,75,150],
       "study_hours":[2,3,4,5,5,6,7,8,8,10]}
df = pd.DataFrame(data)
df
```

```python
print("Mean:",df["marks"].mean())
```

```python
print("Median:",df["marks"].median())
```

```python
print("Mode:")
print(df["marks"].mode())
```

```python
range= df["marks"].max() - df["marks"].min()
print("Range:",range)
```

```python
print("Variance:", df["marks"].var())
```

```python
print("Standard Deviation:", df["marks"].std())
```

```python
print(df["marks"].quantile([0.25,0.50,0.75]))
```

```python
print("90th Percentile")

print(df["marks"].quantile(0.90))
```

```python
Q1 = df["marks"].quantile(0.25)

Q3 = df["marks"].quantile(0.75)

IQR = Q3 - Q1

Outlier_lower = Q1 - 1.5 * IQR

Outlier_upper = Q3 + 1.5 * IQR

outliers = df[(df["marks"] < Outlier_lower) | (df["marks"] > Outlier_upper)]

print(outliers)
```

```python
df.corr()
```

## Business Insights

- Mean represents the average marks.
- Median is less affected by outliers.
- 150 is detected as an outlier using the IQR method.
- Marks and Study Hours show a positive correlation.
- As study hours increase, marks generally increase.

# 🎯Interview Takeaways

1. Mean is affected by outliers.
2. Median is preferred for skewed data.
3. Standard Deviation measures spread.
4. Correlation does not imply causation.
5. IQR is commonly used for outlier detection.
6. Statistics forms the foundation of Data Science.
